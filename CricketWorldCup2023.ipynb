```python
soup.find_all('th')
```

```python
teams_titles = soup.find_all('th')
```

```python
teams_titles
```

```python
teams_table_titles = [title.text.strip() for title in teams_titles]

print(teams_table_titles)
```

```python
filtered_teams_table_titles = [title for title in teams_table_titles if title not in ['PLD', 'NRR']]
```

```python
print(filtered_teams_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = filtered_teams_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #print(individual_row_data)
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'D:\Github\CricketWorldCup2023Prediction\dataFrame/Standings.csv'
df = pd.read_csv(file_path)
```

```python
df.head(10)
```

```python
df.isnull()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Team'], df['PointsPts'])
plt.xlabel('Team')
plt.ylabel('Points')
plt.title('Points for Each Team')
plt.show()
```

```python
plt.scatter(df['Team'], df['PointsPts'])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Team'], df['PointsPts'])
```

```python
winner = df.loc[df['PointsPts'].idxmax(), 'Team']
print(f"The predicted winner of the World Cup is: {winner}")
```

```python
df_sorted = df.sort_values(by='PointsPts', ascending=False)
top_3_teams = df_sorted.head(3)
print("Top 3 Teams:")
print(top_3_teams[['Team', 'PointsPts']])
```
